```python
import matplotlib.pyplot as plt

import KGCN


batch_arguments = [
    # varied random seeds
    '--dataset product --mixer transe --n_epochs 100 --batch_size=16 --random_seed 1',
    # '--dataset product --mixer transe --n_epochs 10 --batch_size=16 --random_seed 2',
    # '--dataset product --mixer transe --n_epochs 10 --batch_size=16 --random_seed 4',
]
```

```python
def plot_result(result):
    # fig, (ax1,ax2) = plt.subplots(2,1, figsize=(10,6))
    fig = plt.figure()
    gs = fig.add_gridspec(2, hspace=0)
    ax1, ax2 = gs.subplots(sharex=True)
    fig.suptitle('Experiment Metics Tracking')
    
    ax1.plot(result[0], label='Train Loss')
    ax1.plot(result[1], label='Evaluation Loss')
    ax2.plot(result[2], label='AUC ROC')

    ax1.set(ylabel='Loss')
    ax1.grid(linestyle = '--', linewidth = 0.5)
    ax2.set(xlabel='Epoch', ylabel='AUC ROC')
    ax2.grid(linestyle = '--', linewidth = 0.5)
    
    plt.tight_layout()
    
    plt.show()
```

```python
results = []
for arg in batch_arguments:
    print(f'[Info] Experiment: {arg}')
    result = KGCN.experiment(arg)
    plot_result(result)
    results.append(result)
    
```
