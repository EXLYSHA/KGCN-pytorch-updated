```python
import matplotlib.pyplot as plt
import numpy as np

import KGCN


batch_arguments = [
    # # varied random seeds
    # '--dataset product --mixer transe --n_epochs 100 --batch_size 16 --random_seed 114',
    # '--dataset product --mixer transe --n_epochs 100 --batch_size 16 --random_seed 123',
    # '--dataset product --mixer transe --n_epochs 100 --batch_size 16 --random_seed 529',
    # test mixer
    '--dataset product --mixer transe --n_epochs 150 --random_seed 114',
    '--dataset product --mixer attention --n_epochs 150 --random_seed 114',
    '--dataset product --mixer transe --n_epochs 150 --random_seed 222',
    '--dataset product --mixer attention --n_epochs 150 --random_seed 222',
    '--dataset product --mixer transe --n_epochs 150 --batch_size 16 --random_seed 234',
    '--dataset product --mixer attention --n_epochs 150 --batch_size 16 --random_seed 234',
    '--dataset product --mixer transe --n_epochs 150 --batch_size 16 --random_seed 134',
    '--dataset product --mixer attention --n_epochs 150 --batch_size 16 --random_seed 134',
]
```

```python
def plot_result(result):
    """
    :param result: A tuple, (loss_list, test_loss_list, auc_score_list)
    """
    fig = plt.figure()
    gs = fig.add_gridspec(2, hspace=0)
    ax1, ax2 = gs.subplots(sharex=True)
    fig.suptitle('Experiment Metics Tracking')
    
    ax1.plot(result[0], label='Train Loss')
    ax1.plot(result[1], label='Evaluation Loss')
    ax2.plot(result[2], label='ROC AUC')

    ax1.set(ylabel='Loss')
    ax1.grid(linestyle = '--', linewidth = 0.5)
    ax2.set(xlabel='Epoch', ylabel='ROC AUC')
    ax2.grid(linestyle = '--', linewidth = 0.5)
    
    plt.tight_layout()
    
    plt.show()

def print_result_summary(result):
    train = np.array(result[0])
    test = np.array(result[1])
    auc = np.array(result[2])

    i = np.argmin(train)
    j = np.argmin(test)
    k = np.argmax(auc)
    print('[Result] (epoch, train loss, test loss, auc)')
    print(f'│ Min train loss: {i:>3}, {train[i]:2.6f}, {test[i]:2.6f}, {auc[i]:2.6f}')
    print(f'│  Min test loss: {j:>3}, {train[j]:2.6f}, {test[j]:2.6f}, {auc[j]:2.6f}')
    print(f'└    MAX ROC AUC: {k:>3}, {train[k]:2.6f}, {test[k]:2.6f}, {auc[k]:2.6f}')
```

```python
# Show result
for arg in batch_arguments:
    print(f'[Info] Experiment: {arg}')
    result = KGCN.experiment(arg)
    plot_result(result)
    print_result_summary(result)
    print()

    # Remember to check the convergence speed, performance at convergence, best AUC
```
